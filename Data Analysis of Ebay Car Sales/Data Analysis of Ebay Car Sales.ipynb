Based on our observations: 
* We will drop the following columns: `seller`,`offer_type` and `num_photos`

```python
autos.drop(['offer_type','seller','num_photos'],axis=1)
```

* We will convert the columns `price` and `odometer` from string values to numeric ones

```python
autos['price']=(autos['price']
                        .str.replace('$','')
                        .str.replace(',','')
                        .astype(int))
autos['odometer']=(autos['odometer']
                            .str.replace('km','')
                            .str.replace(',','')
                            .astype(int))
```

```python
autos.rename(columns={'odometer':'odometer_km'},inplace=True)
```

### Exploring Price by Brand

```python
autos=autos[autos['registration_year'].between(1900,2016)]
```

```python
autos['registration_year'].describe()
```

We narrowed the values from 48565 to 46681. We removed 1884 entries.

```python
autos['brand'].value_counts(normalize=True)
```

```python
autos_brand=autos['brand'].value_counts(normalize=True)
top_brands=autos_brand[autos_brand > .05].index #index is important for the next step
print(top_brands)
```

We are exploring the top car companies that are listed on eBay Germany. The 50% of the market share is made up from German companies, with VW leading the chart 
with 21%.

```python
top_autos_price={}
for c in top_brands:
    selected_brands=autos[autos['brand']==c]
    means=selected_brands['price'].mean()
    top_autos_price[c]=int(means)
top_autos_price
```


Of the top 5 brands, there is a distinct price gap:

Audi, BMW and Mercedes Benz are more expensive
Ford and Opel are less expensive
Volkswagen is in between - this may explain its popularity, it may be a 'best of 'both worlds' option.

### Exploring Aggregate Data in DataFrame

```python
top_mean_mileage={}
for c in top_brands:
    selected_brands=autos[autos['brand']==c]
    means=selected_brands['odometer_km'].mean()
    top_mean_mileage[c]=means
top_mean_mileage
```

```python
tap_series = pd.Series(top_autos_price)
tap=pd.DataFrame(tap_series, columns=["mean_price"])
tap
```

```python
tmm_series = pd.Series(top_mean_mileage)
tmm=pd.DataFrame(tmm_series, columns=["mean_mileage"])
tmm
```

```python
tmm["mean_price"] = tap
```

```python
tmm
```


The range of car mileages does not vary as much as the prices do by brand, instead all falling within 10% for the top brands. There is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage.
